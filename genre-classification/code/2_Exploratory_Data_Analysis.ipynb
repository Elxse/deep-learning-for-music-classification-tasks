# [Genre] Exploratory Data Analysis

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
plt.style.use('ggplot')
```

```python
# Creating a list of genres
genres = ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop','reggae', 'rock']
```

```python
DATA_PATH = '../data/'
AUDIO_ROOT_PATH = DATA_PATH + 'genre-audios/'
MELSPECS_FOLDER = DATA_PATH + 'genre-melspecs-19000/'
SPECS_FOLDER = DATA_PATH + 'genre-specs-10000/'
MFCCS_FOLDER = DATA_PATH + 'genre-mfccs-18996/'
```

# <a name='spec'> </a> 1. Plotting Spectrograms

```python
def spectrogram_subplots(genres, n_fft=2048, hop_length=512):
    """Plots a mel spectrogram for one song per genre in a 5 x 2 grid."""
    
    # Defining the subplots
    fig, ax = plt.subplots(nrows = 2, ncols = 5, figsize = (25,10))
    ax = ax.ravel()

    # Looping through the list of genres
    for i, kind in enumerate(genres):
        
        # Reading the first file from each genre
        spec = np.load(SPECS_FOLDER + f'{kind}.00000.0.spec.npy')
        
        # Displaying the spectrogram 
        librosa.display.specshow(spec, x_axis='time', y_axis='log', ax=ax[i])
        ax[i].set_title(str(kind))
```

```python
# Plotting mel spectrograms for each genre
spectrogram_subplots(genres)
```

# <a name='melspec'> </a> 2. Plotting Mel Spectrograms

```python
def melspectrogram_subplots(genres):
    """Plots a mel spectrogram for one song per genre in a 5 x 2 grid."""
    
    # Defining the subplots
    fig, ax = plt.subplots(nrows = 2, ncols = 5, figsize = (25,10))
    ax = ax.ravel()

    # Looping through the list of genres
    for i, kind in enumerate(genres):
        
        # Reading the first file from each genre
        melspec = np.load(MELSPECS_FOLDER + f'{kind}.00000.0.melspec.npy')
        
        # Displaying the mel spectrogram 
        librosa.display.specshow(melspec, y_axis='mel', x_axis='time', ax=ax[i])
        ax[i].set_title(str(kind))
```

```python
# Plotting mel spectrograms for each genre
melspectrogram_subplots(genres)
```

# <a name='mfcc'> </a> 3. Plotting MFCCs

```python
def mfcc_subplots(genres):
    """Plots MFCCs for one song per genre in a 5 x 2 grid."""
    
    # Defining the subplots
    fig, ax = plt.subplots(nrows = 2, ncols = 5, figsize = (25,10))
    ax = ax.ravel()

    # Looping through the list of genres
    for i, kind in enumerate(genres):
        
        # Reading the first file from each genre
        mfccs = np.load(MFCCS_FOLDER + f'{kind}.00000.0.mfcc.npy')
        
        # Displaying MFCCs
        librosa.display.specshow(mfccs, x_axis='time', ax=ax[i])
        ax[i].set_title(str(kind))
```

```python
mfcc_subplots(genres)
```

Samples above show that the spectrograms, mel spectrograms and MFCCs are different for different genres.
